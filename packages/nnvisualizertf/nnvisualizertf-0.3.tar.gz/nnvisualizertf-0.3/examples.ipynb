```python
import src.nnvisualizertf as nn_visualizer

from tensorflow.keras.models import Model
from tensorflow.keras.layers import AveragePooling2D, Conv2D, Dense, Dropout, Flatten, Input, concatenate
```

```python
# define exemplary network
in_layer = Input(shape=(40, 40, 1))
conv1 = Conv2D(16, 3, strides=1, activation='relu')(in_layer)
pool1 = AveragePooling2D(2, strides=2)(conv1)
conv2 = Conv2D(32, 3, strides=1, activation='relu')(pool1)
pool2 = AveragePooling2D(2, strides=2)(conv2)
flat = Flatten()(pool2)
dense1 = Dense(512)(flat)

model = Model(inputs=in_layer, outputs=dense1)
```

```python
nn_visualizer.visualize_model(model)
```
